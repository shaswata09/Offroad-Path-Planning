```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import cProfile, pstats, sys
import cv2
import sys
import time
from matplotlib import pyplot as plt

import timeit
import random
import math
import os
import tcod
import copy
from os.path import isfile,join
cv2.namedWindow('image')



img = None
img2 = None
randarray = np.linspace(0, np.pi, 50)



def mouse_evt(event, x, y, flags, param):
    global img, img2
    # Mouse is Moving
    if event == cv2.EVENT_MOUSEMOVE and np.array_equal(img[y][x],[255,255,255]):
        img_copy = img.copy()
        cv2.circle(img_copy,(x,y),1,(255,0,0),-1)
        castRays(x,y,img_copy)
        cv2.imshow('image',img_copy)
    else:
        cv2.imshow('image',img)


def castRays(x,y, img_Copy):
    global img
    global img2
    final_dist = 500
    current_point = (x,y)
    for xr in randarray:
        x1 = int(x + final_dist * np.cos(xr))
        y1 = int(y + final_dist * np.sin(xr))
        if (lop := bresenham((x,y), (x1,y1), img)) == True:
            cv2.line(img_Copy,(x,y),(x1,y1),(0,255,0),1)
        else:
            cv2.line(img_Copy, (x,y), lop[1], (0,0,255),1)

            
        
    return 




    
def bresenham(start, end, img_c):
    """
    Bresenham's Line Generation Algorithm
    https://www.youtube.com/watch?v=76gp2IAazV4
    """
    global img
    global img2

    if start[0] < 0 or start[1] < 0 or start[1] >= len(img) or start[0] >= len(img) or np.array_equal(img[start[1]][start[0]], [0,0,0]):
        img[start[1]][start[0]] = img2[start[1]][start[0]]
        return False, (start[0],start[1])
        
    (x0, y0) = start
    (x1, y1) = end
    
    
    m = None

    
    dx = abs(x1 - x0)
    dy = abs(y1 - y0)
    if dx == 0:
        m = 1.0
    else:
        m = dy/dx
        
    mm = False
    # step 3 perform test to check if pk < 0
    flag = True
    
    line_pixel = []

    line_pixel.append((x0,y0))
    
    step = 1
    if x0>x1 or y0>y1:
        step = -1
        
    if m < 1:
        x0, x1 ,y0 ,y1 = y0, y1, x0, x1
        dx = abs(x1 - x0)
        dy = abs(y1 - y0)
        mm = True
        
    p0 = 2*dx - dy
    x = x0
    y = y0
    term = None
    for i in range(abs(y1-y0)):
        if flag:
            x_previous = x0
            p_previous = p0
            p = p0
            flag = False
        else:
            x_previous = x
            p_previous = p
        
    
        if p >= 0:
            x = x + step

        p = p_previous + 2*dx -2*dy*(abs(x-x_previous))
        y = y + step
        
        if mm:
            if x < 0 or y < 0 or y >= len(img) or x >= len(img) or  np.array_equal(img[x][y],[0,0,0]):
                img[x][y] = img2[x][y]
                return False, line_pixel[-1]
            img[x][y] = img2[x][y]
            line_pixel.append((y,x))
        else:
            if x < 0 or x >= len(img) or y < 0 or y >= len(img) or np.array_equal(img[y][x],[0,0,0]):
                img[y][x] = img2[y][x]
                return False, line_pixel[-1]
            img[y][x] = img2[y][x]
            line_pixel.append((x,y))
            
    return True
     
            
def grabRandomImage():
    #Returns an image path to use for the path processing

    imgFolder = "C:/Users/charles/Downloads/sample_predictions2/predictions"
    groundTruthFolder = "C:/Users/charles/Downloads/sample_predictions2/ground_truth"

    imgs = [f for f in os.listdir(imgFolder) if isfile(join(imgFolder,f))]

    return join(imgFolder, imgs[10]), join(groundTruthFolder, imgs[10])

#orientate towards the next point in the path. 

def main():
    os.environ["CUDA_VISIBLE_DEVICES"] = "1"
    global img 
    global img2
    imgPath, imgPathgroundTruth = grabRandomImage()
    img = cv2.imread(imgPath)
    img2 = cv2.imread(imgPathgroundTruth)
    
    
    cv2.setMouseCallback("image", mouse_evt)

    cv2.imshow('image',img)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
#
if __name__ == "__main__":
    main()
```
